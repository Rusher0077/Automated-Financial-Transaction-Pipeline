```python

from datetime import date, timedelta
from pyspark.sql.functions import col, lit, expr

yesterday = date.today() - timedelta(days=1)

df_raw = spark.table("pay_slim")

df_offset = df_raw.withColumn(
    "day_number", ((col("step") - 1) / 24).cast("int") + 1
).withColumn(
    "simulated_date", lit(yesterday) + expr("INTERVAL 1 DAY") * (col("day_number") - 1)
)

df_offset.write.format("delta").mode("overwrite").saveAsTable("transactions_with_dates")
```
